```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv("./calibration_data_for_debug_more_data.csv")
df.plot(y=['plat_to_world_roll'])
print(df['plat_to_world_pitch'].mean())
print(df['plat_to_world_roll'].mean())
```

```python
filtered_for_good_readings = df.where(df['ncorners_world'] + df['ncorners_platform'] == 32)

tailed = filtered_for_good_readings.drop(filtered_for_good_readings.tail(100).index)
tailed = filtered_for_good_readings[(np.abs(stats.zscore(df['plat_to_world_pitch'])) < 3)]
#tailed = filtered_for_good_readings
print(tailed['plat_to_world_pitch'].mean())
print(tailed['plat_to_world_roll'].mean())
tailed.plot(y=['plat_to_world_roll', 'plat_to_world_pitch'])
```

```python
df_manual_slot_moves = pd.read_csv('locations_manual_slot_moves.csv')
import matplotlib.pyplot as plt
from numpy.polynomial import Polynomial

def smooth(data, nvals):
    output = []
    
    for i in range(nvals, len(data)):
        x1 = data[i-1]
        x2 = data[i]
        output.append(
            ((x1 + x2) / 2.0)
        )
        
    return output

data = smooth(df_manual_slot_moves['z'])
nrows = len(data)
xvals = list(map(float, range(len(data))))
plt.plot(data)
poly = Polynomial.fit(x=xvals, y=data, deg=1)
plt.plot(xvals, list(map(poly, xvals)))
```
